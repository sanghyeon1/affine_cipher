```python
# 아핀 암호 공격.
import math


def affine_cryp():
    z_26 = [chr(i) for i in range(97, 123)]
    plainText = input("Input plaintext: ")
    cryptogram = input("Input cryptogram: ")

    # 평문, 암호문 문자열 리스트화 후 숫자 변환.
    pl_num_list, cr_num_list = get_text_to_num_list(z_26, plainText, cryptogram)

    # ax + k 에 해당되는 암호키 K(a, k) 해독.
    a, k = calculate_mod(pl_num_list, cr_num_list)

    idx = (a * pl_num_list[0] + k) % 26
    if idx == cr_num_list[0]:
        print(f"암호키: K({a}, {k})")


def get_text_to_num_list(z_26, plainText, cryptogram):
    plainText_list = list(plainText.replace(" ", ""))
    cryptogram_list = list(cryptogram.replace(" ", ""))

    pl_num_list, cr_num_list = [], []
    for i in range(len(plainText_list)):
        num1 = z_26.index(plainText_list[i])
        num2 = z_26.index(cryptogram_list[i])
        pl_num_list.append(num1)
        cr_num_list.append(num2)

    return pl_num_list, cr_num_list


def calculate_mod(pl_num_list, cr_num_list):
    # "a*x + k = y mod 26"을 위한 변수 생성.
    x1 = pl_num_list[0]
    x2 = pl_num_list[1]
    y1 = cr_num_list[0]
    y2 = cr_num_list[1]

    if x1 > x2:
        x = x1 - x2
        y = y1 - y2
    else:
        x = x2 - x1
        y = y2 - y1

    if y < 0:
        y = 26 + y

    # ax + k에서 a, k 키 값 찾는 함수.
    a_candidates = find_possible_a(x, y)
    a = a_candidates[0]
    k_candidates = find_possible_k(a, x1, y1)
    k = k_candidates[0]

    return a, k


def find_possible_a(x, y):
    a_list = []
    for i in range(12):
        a_list.append((y + 26 * i) / x)
    # print(a_list)
    candidates = []
    for i in range(12):
        if a_list[i].is_integer() and a_list[i] < 26:
            if math.gcd(int(a_list[i]), 26):
                candidates.append(int(a_list[i]))
    # print(candidates)
    return candidates


def find_possible_k(a, x1, y1):
    k_list = []
    for i in range(26):
        k_list.append((y1 + 26 * i) - (a * x1))
    # print(k_list)
    candidates = []
    for i in range(26):
        if 1 <= k_list[i] <= 26:
            candidates.append(int(k_list[i]))
    # print(candidates)
    return candidates



def get_cryptogram():
    z_26 = [chr(i) for i in range(97, 123)]

    plainText = input("Input plaintext: ")
    plainText_list = list(plainText.replace(" ", ""))
    a, k = map(int, input("Input key a and k: ").split())

    pl_num_list, changed_num = [], []
    for i in range(len(plainText_list)):
        num1 = z_26.index(plainText_list[i])

        pl_num_list.append(num1)
        if num1*a+k > 26:
            t = (num1 * a + k) % 26
        else:
            t = num1*a+k

        changed_num.append(t)

    temp = []
    for i in changed_num:
        temp.append(z_26[i])
    result = ''.join(temp)

    print("cryptogram: ", result)
    # print(pl_num_list, changed_num)


if __name__=="__main__":
    order = int(input("get_cryptogram: 1 // affin_cryp: 2 --> "))
    while 1:
        if order == 1:
            get_cryptogram()
            break
        elif order == 2:
            affine_cryp()
            break
        else:
            order = int(input("Wrong number. input 1 or 2: "))
```
